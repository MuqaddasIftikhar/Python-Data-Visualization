```python
import seaborn as sns 
import matplotlib.pyplot as plt 
tip=sns.load_dataset("tips")
tip
sns.boxplot(x="time", y="tip", data=tip, showmeans=True, meanprops={"marker":"*", "markersize":"1", "markeredgecolor":"1"},)
plt.title("BOX PLOT OF TIPS DATASET", size=15)
plt.xlabel("Time (Restaurant)", size=15)
plt.ylabel("Tips (Restaurant)", size=15)
# Adding facetgrid function
sns.FacetGrid(tip, col="time", row="smoker")
# Adding Facet Wrap Function
plt.face
plt.show()
```
