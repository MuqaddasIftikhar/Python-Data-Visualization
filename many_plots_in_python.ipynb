```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
dot=sns.load_dataset("dots")
dot
# Defining a color palette
palette=sns.color_palette('rocket_r')
# ploting
sns.relplot(data=dot, x="time", y="firing_rate", hue="coherence", size="choice", col="align", kind="line", 
            size_order=["T1", "T2"], palette=palette, height=5, aspect=0.75, facet_kws=dict(sharex=False),)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set_theme(style="ticks", palette="pastel")
tip=sns.load_dataset("tips")
tip
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill", data=tip, hue="smoker", palette=["m", "g"])
sns.despine(offset=10, trim=True)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
sns.set_theme(style="whitegrid")
tip=sns.load_dataset("tips")
tip
sns.violinplot(data=tip, x="day", y="total_bill", hue="smoker", split=True, inner="quart", linewidth=1, palette={"Yes":"b", "No":"0.85"})
sns.despine(left=True)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set_theme(style="whitegrid")
diamonds=sns.load_dataset("diamonds")
f, ax=plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking=["I1", "SI2", "SI1", "VS2", "VS1", "WS2", "WS1", "IF"]
sns.scatterplot(x="carat", y="price", hue="clarity", size="depth", palette="ch:r=-0.2, d=0.3_r", hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0, data=diamonds, ax=ax)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load dataset
diamonds=sns.load_dataset("diamonds")
diamonds
# Draw a line plot 
sns.lineplot(x="carat", y="price", data=diamonds)
plt.show()
```
