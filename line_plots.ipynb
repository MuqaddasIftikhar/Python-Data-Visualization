6. Adding Tiltes

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load Dataset
flower=sns.load_dataset("iris")
flower
# Draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("FLOWERS PLOT")
plt.show()
```

7. Adding Limits

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load Dataset
flower=sns.load_dataset("iris")
flower
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
# Adding Titles
plt.title("FLOWERS PLOT")
#  Adding Limits
plt.xlim(2)
plt.ylim(1)
plt.show()
```

8. Adding Labels


```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load Dataset
flower=sns.load_dataset("iris")
flower
# Draw a line plot 
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
# Adding Titles
plt.title("FLOWERS PLOT")
# Adding Limits
plt.xlim(2)
plt.ylim(1)
# Adding labels
plt.xlabel("plot on x-axis")
plt.ylabel("plot on y-axis")
plt.show()
```

9. change the figure

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load Dataset
flower=sns.load_dataset("iris")
flower
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
# Adding titles
plt.title("FLOWER PLOT OF IRIS DATASET")
# Adding limits
plt.xlim(2)
plt.ylim(1)
# Adding labels
plt.xlabel("plot on x-axis")
plt.ylabel("plot on y-axis")
# Change figure
plt.figure()
plt.show()
```

10. Adding set Styles

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load Dataset
flower=sns.load_dataset("iris")
flower
# Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
# Adding titles
plt.title("FLOWER PLOT OF IRIS DATASET")
# Adding limits
plt.xlim(2)
plt.ylim(1)
# Adding labels
plt.xlabel("plot on x-axis")
plt.ylabel("plot on y-axis")
# Change figure
plt.figure()
# Adding set Styles
sns.set_style("darkgrid")
plt.show()
```
