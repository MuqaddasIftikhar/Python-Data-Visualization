```python
import seaborn as sns 
import matplotlib.pyplot as plt 
# Load dataset
flower=sns.load_dataset("iris")
flower
# Draw a bar plot 
sns.barplot(x="sepal_length", y="sepal_width", data=flower)
# Adding titles
plt.title("DATASET OF IRIS PLOT")
# Adding limits
plt.xlim(2)
plt.ylim(1)
# Adding labels
plt.xlabel("plot on x-axis")
plt.ylabel("plot on y-axis")
# Adding figures
plt.figure(2)
# Adding set styles
sns.set_style("ticks")
plt.show()
```

# Boat Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt 
# Load Dataset
boat=sns.load_dataset("titanic")
boat
# Draw a bar plot 
sns.barplot(x="survived", y="age", hue="who", data=boat)
# Adding titles
plt.title("PLOT OF TITANIC DATASET")
plt.show()
```
